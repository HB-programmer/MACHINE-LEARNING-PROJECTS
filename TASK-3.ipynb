## INDIVIDUAL VISUALIZATION OF THE CATEGORIES

### NOW WE PLOT A GRAPH SHOWING THE TERRORISM COUNT IN DIFFERENT AREAS

```python
plt.figure(figsize=(25,18))
sns.countplot(x=df["region_txt"])
print(df["region_txt"].value_counts())
```

 ### SO FROM ABOVE FIGURE WE CAN ESTIMATE THAT MIDDLE EAST AND NORTH AFRICA ARE MOST 
 
 ### AFFECTED BY TERRORISM AND AFTER AFRICA SOUTH ASIA IS SECOND MOST AFFECTED AREA

## NOW WE PLOT A GRAPH WHICH SHOWS THE NUMBER OF TERRORISM ATTACKS ON DIFFERENT COUNTRIES

```python
plt.figure(figsize=(25,18))
sns.countplot(x=df["country_txt"])
print(df["country_txt"].value_counts())
```

## NOW WE PLOT A BAR CHART WHICH SHOWS THE TOP 10 COUNTRIES AFFECTED BY TERRORISM

```python
df["country_txt"].value_counts().head(10).plot.bar()
```

### SO FROM ABOVE BAR CHART WE CAN SEE THAT THE MOST TERRORISM ACTIVITES HAS OCCURED IN IRAQ 
### FOLLOWED BY PKAISTAN AND AFGHANISTAN

### NOW WE PLOT A PIE CHART TO SHOW THE ABOVE RESULT

```python
df["country_txt"].value_counts().head(10).plot.pie()
```

#### FROM ABOVE FIGURE WE CAN ESTIMATE THAT MOST OF THE TERRORISM ACTIVITIES HAVE OCCURED IN IRAQ FOLLOWED BY PAKISTAN
#### AFGHANISTAN AND INDIA

## THEREFORE, WE CAN SAY IRAQ IS THE HOT ZONE OF TERRORISM

## NOW WE PLOT A COUNTPLOT SHOWING THE DETAILS OF ATTACKTYPE WITH THEIR NUMBER OF USAGE TIME

```python
plt.figure(figsize=(18,13))
df["attacktype1_txt"].value_counts().head(10).plot.pie()
```

### FROM ABOVE PIE CHART WE CAN SEE THAT TERRORIS MOSTLY USE EXPLOSIONS FOR THEIR EVIL ACTIVITIES

## NOW WE DRAW A PIE CHART WHICH SHOWS WHICH TYPE OF TAGET , TARGETED BY TERRORIST

```python
plt.figure(figsize=(18,13))
df["targtype1_txt"].value_counts().head(10).plot.pie()
```

### FROM ABOVE PIE CHART WE CAN SEE THAT TERRORIS MOSTLY TARGET PRIVATE CITIZENS AND PROPERTIES FOLLOWED BY MILITARY AND POLICE

## NOW WE DRAW THE BARPLOT WHICH SHOWS THAT IN WHICH YEAR MOST OF TERRORIST ACTIVITIES HAD OCCUR

```python
plt.figure(figsize=(18,12))
df["iyear"].value_counts().head(10).plot.bar()
```

### FROM ABOVE BAR PLOT WE CAN SEE THAT THE MOST OF THE TERRORIST ACTIVITIES HAD OCCURED IN 2014

## NOW WE DRAW A VERTICAL BAR CHART TO SHOW THAT WHICH TERRORIST GROUP HAD PERFORMED THE MOST OF THE EVIL ACTIVITIES

```python
plt.figure(figsize=(25,18))
sns.barplot(x=df["gname"].value_counts()[1:11],y=df["gname"].value_counts()[1:11].index,palette='viridis')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```
